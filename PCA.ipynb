```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("forestfires.csv")
data
```

```python
features = data.drop(["X","Y","area"], axis=1)
target = data["area"]
```

```python
features.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
features['month'] = label_encoder.fit_transform(features['month'])
features['day'] = label_encoder.fit_transform(features['day'])
features.head()
```

```python
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
pca = PCA(n_components = 3)
principalComponents = pca.fit_transform(features)
principalDataframe = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2','PC3'])
principalComponents
```

```python
newDataframe = pd.concat([principalDataframe, target],axis = 1)
newDataframe
```
